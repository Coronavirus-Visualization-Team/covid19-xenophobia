```
data_type = {"date": "string",	"id": "string",	"coordinates": "string", "full_text":"string", "sentiment": "string",	"key_words":"string", "hashtags":"string"}
data = pd.read_csv("trainingdata.csv", dtype=data_type)

data_length = len(data)
data_training = round(data_length*0.7)

X_train = data["full_text"][:data_training]
X_val = data["full_text"][data_training:] #add separate CSV if not completed
y_val = data["sentiment"][data_training:]
y_train = data["sentiment"][:data_training]
```

```
tfidf = TfidfVectorizer(ngram_range=(1,2), max_df= 0.25, min_df= 0.01) #experiment with parameters of max_df and min_df (max and min df can be between 0.01 and 0.99) (i dont recommend experimenting with ngram range but you can look up the documentation and experiment if you want). NOTE- WHILE EXPERIMENTING, YOU CANNOT TRY TO MAXIMIZE TRAINING ACCURACY, ONLY TRY TO MAXIMIZE VAL ACCURACY BY VARYING PARAMETERS. (try to avoid overfitting)
X_train_tfidf = tfidf.fit_transform(X_train) 
X_val_tfidf = tfidf.transform(X_val)
```

```
from sklearn.naive_bayes import MultinomialNB
nb_body = MultinomialNB() 
nb_body.fit(X_train_tfidf, y_train) 
y_train_pred = nb_body.predict(X_train_tfidf)
print('Naive Bayes In train F1 and Accuracy Scores:')
print('F1 score {:.4}%'.format(f1_score(y_train, y_train_pred, average='macro')*100 ))
print ('Accuracy score {:.4}%'.format(accuracy_score(y_train, y_train_pred)*100))
y_val_pred = nb_body.predict(X_val_tfidf)
print('Naive Bayes In val Accuracy and other metrics:')
print('F1 score {:.4}%'.format(f1_score(y_val, y_val_pred, average='macro')*100 ))
print ('Accuracy score {:.4}%'.format(accuracy_score(y_val, y_val_pred)*100))
```
